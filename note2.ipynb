Web Scraping Python

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
```

เปิดเว็บไซต์

```python
driver = webdriver.Chrome()
driver.get('https://www.set.or.th/th/market/index/set100/overview')
```

ดึงข้อมูล

```python
data = driver.page_source # ดึงโครงสร้าง html selenium
datadf = pd.read_html(data)[2] # ดึงข้อมูลจาก html pandas
datadf
```

```python
newcolumn = []

for i in datadf.columns:
    newcolumn.append(i.replace("  (Click to sort Ascending)", "")) #ลบคำนี้
#newcolumn

datadf.columns = newcolumn #เปลี่ยน column ใหม่
datadf
```

```python
data2 = datadf
datadf = datadf.set_index("หลักทรัพย์") #ตัวแทนเรียกใช้
```

สร้าง url หุ้นแต่ละตัว

```python
#stock = datadf['หลักทรัพย์'][0]
stock = datadf.index[0]

url = 'https://www.set.or.th/th/market/product/stock/quote/'+ stock +'/price' #ใส่ชื่อหุ้นในลิ้ง

for i in datadf.index:
    url = 'https://www.set.or.th/th/market/product/stock/quote/'+ i +'/price'
    print(url)

for i in datadf.index[0:5]:
    url = 'https://www.set.or.th/th/market/product/stock/quote/'+ i +'/price' #เปิดหลายหน้า
    driver.get(url)
    
```
